````python
from sklearn.datasets import load_iris

iris = load_iris(as_frame=True)
df = iris.frame
df.head()
````

## 2. Inspeção inicial

````python
print(f'Shape: {df.shape}')
print(f'Tipos:\n{df.dtypes}')
print(f'\nNulos por coluna:\n{df.isnull().sum()}')
print(f'\nDuplicados: {df.duplicated().sum()}')
````

## 3–6. Estatísticas descritivas

````python
df.describe().T
````

## 7. Detecção de outliers (IQR)

````python
numeric_cols = df.select_dtypes(include='number').columns
for col in numeric_cols:
    q1, q3 = df[col].quantile([0.25, 0.75])
    iqr = q3 - q1
    outliers = df[(df[col] < q1 - 1.5 * iqr) | (df[col] > q3 + 1.5 * iqr)]
    print(f'{col}: {len(outliers)} outlier(s)')
````

## 8. Visualizações

````python
fig, axes = plt.subplots(1, len(numeric_cols), figsize=(16, 4))
for ax, col in zip(axes, numeric_cols):
    df[col].plot.hist(ax=ax, bins=20, edgecolor='white')
    ax.set_title(col)
plt.suptitle('Histogramas — Iris', y=1.02)
plt.tight_layout()
plt.show()
````

````python
fig, ax = plt.subplots(figsize=(10, 4))
df[numeric_cols].boxplot(ax=ax)
ax.set_title('Boxplot — Iris')
plt.tight_layout()
plt.show()
````

````python
plt.figure(figsize=(7, 5))
sns.heatmap(df[numeric_cols].corr(), annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Heatmap de Correlação — Iris')
plt.tight_layout()
plt.show()
````
