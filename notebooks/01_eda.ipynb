# 01 — Análise Exploratória e Estatística Descritiva

**Objetivo**: explorar um dataset com a biblioteca `ml-clustering-lab`, calculando medidas
de tendência central, dispersão, forma e qualidade dos dados, além de gerar visualizações
univariadas e bivariadas.

## Tópicos
1. Carregamento de dados
2. Inspeção inicial (shape, tipos, nulos, duplicados)
3. Medidas de tendência central (média, mediana, moda)
4. Medidas de dispersão (variância, desvio padrão, IQR, amplitude)
5. Medidas de posição (quartis, percentis)
6. Forma da distribuição (assimetria, curtose)
7. Detecção de outliers (IQR e z-score)
8. Visualizações: histogramas, boxplots, heatmap de correlação

```python
# Imports do projeto ml-clustering-lab
from ml_clustering_lab.datasets.loaders import load_sklearn
from ml_clustering_lab.stats.descriptive import (
    describe_dataframe,
    central_tendency,
    dispersion,
    position_measures,
    shape_measures,
    detect_outliers,
)
from ml_clustering_lab.visualization.plots import (
    plot_histogram,
    plot_boxplot,
    plot_correlation,
)

import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

## 1. Carregamento de dados

Usaremos o dataset **Iris** como exemplo.
O módulo `datasets.loaders` suporta: `iris`, `wine`, `breast_cancer`, `digits`,
`blobs`, `moons`, `circles` e qualquer CSV local ou URL remota.

```python
df = load_sklearn('iris')
print(f'Dataset carregado: {df.shape[0]} linhas × {df.shape[1]} colunas')
df.head()
```

## 2. Inspeção inicial

```python
summary = describe_dataframe(df)
print(f"Shape: {summary['shape']}")
print(f"Nulos por coluna: {summary['null_counts']}")
print(f"Duplicados: {summary['duplicates']}")
print()
summary['numeric_summary']
```

## 3. Medidas de tendência central

```python
numeric_cols = df.select_dtypes(include='number').columns.tolist()

for col in numeric_cols:
    ct = central_tendency(df[col])
    print(f'{col:30s} | média={ct["mean"]:.3f}  mediana={ct["median"]:.3f}  moda={ct["mode"]:.3f}')
```

## 4. Medidas de dispersão

```python
rows = []
for col in numeric_cols:
    d = dispersion(df[col])
    d['feature'] = col
    rows.append(d)

pd.DataFrame(rows).set_index('feature')
```

## 5. Medidas de posição (quartis e percentis)

```python
for col in numeric_cols:
    pm = position_measures(df[col], percentiles=[0.10, 0.25, 0.50, 0.75, 0.90])
    print(f'{col}: Q1={pm["q1"]:.3f}  Q2={pm["q2"]:.3f}  Q3={pm["q3"]:.3f}  '
          f'P10={pm["p10"]:.3f}  P90={pm["p90"]:.3f}')
```

## 6. Forma da distribuição (assimetria e curtose)

```python
for col in numeric_cols:
    sm = shape_measures(df[col])
    print(f'{col:30s} | assimetria={sm["skewness"]:+.3f}  curtose={sm["kurtosis"]:+.3f}')
```

## 7. Detecção de outliers

```python
for col in numeric_cols:
    r_iqr = detect_outliers(df[col], method='iqr')
    r_z   = detect_outliers(df[col], method='zscore')
    print(f'{col:30s} | IQR: {r_iqr["n_outliers"]} outlier(s) ({r_iqr["pct_outliers"]:.1f}%)'
          f'  |  Z-score: {r_z["n_outliers"]} outlier(s)')
```

## 8. Visualizações

```python
plot_histogram(df)
```

```python
plot_boxplot(df)
```

```python
plot_correlation(df)
```
